```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML, Image
%matplotlib inline
```

```python
# plot
def plot():
    plt.plot(x,y)
    style()
    plt.show()
def scatter():
    style()
    plt.scatter(x, y)
    plt.show()
    
def style():
    plt.axvline(x=0, color = '0.5')  # draw x =0 axes
    plt.axhline(y=0, color = '0.5')   # draw y =0 axes
```

```python
# function   
def f(x):
    return (x+4)**2

# 평균변화율
# y변화량 / x 변화량
def ac(a,b):
    return (f(b)-f(a))/(b-a)
# 순간변화율
# a->x d->극한값 대체
def mc(a,d=0.000001):
    return ac(a,a+d)
x = np.arange(-50,50,0.01)
y = f(x)
plot()
```

```python
#xx = np.random.rand(1)*50-25
xx = 43
# http://www.wolframalpha.com/widgets/view.jsp?id=a8c69548205e7958458742cc1cec9a22
# d\/dx((x + 4)^2) = 2 (x + 4)
alpha = 0.02

fig = plt.figure() 
fig.canvas.draw()
for i in range(100):
    xx += -2*(xx+4)*alpha
    if i%20 == 0:
        plt.subplot(222)
        plt.scatter(xx,f(xx))
        plot()
    #print("%d: x:%f y:%f"%(i,xx,f(xx)))
plt.subplot(111)
plt.scatter(xx,f(xx))
plot()
```

```python
a = 16.88
b = 20.2
print("%d 와 %d 의 평균변화율: %f" % (a, b, ac(a,b)))
print("%d 의 순간변화율: %f" % (a, mc(a)))
print("%d 의 순간변화율: %f" % (b, mc(b)))

plt.scatter(a,f(a))
plt.scatter(b,f(b))
plot()
```
